**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
z = x ** 3

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# Plot on main axis
ax1.plot(x, y)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Main Plot')

# Plot on inset axis
ax2.plot(x, z)
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_title('Inset Plot')

# Set x and y limits on inset plot
ax2.set_xlim([0, 2])
ax2.set_ylim([0, 20])

plt.show()
```
